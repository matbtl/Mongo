```python
import pandas as pd
import pymongo
```

```python
client = pymongo.MongoClient()
database = client['exercices']
collection = database['youtube']
```

```python
df_youtube = pd.read_csv("./data/USvideos.csv")
df_youtube.head()
```

## Question 0

### Netoyer les données

```python
df_youtube["date"]=pd.to_datetime(df_youtube["date"])
# car vide_id n'est pas un id ici
df_youtube["tags"] = df_youtube["tags"].apply(lambda x: x.split("|"))
df_youtube.dtypes
```

```python
df_youtube.head()
```

### Importer les données

```python
collection.delete_many({})
collection.insert_many(df_youtube.to_dict(orient='records'))
```

## Question 1  

```python
cursor = collection.find({"channel_title":"Apple"})
pd.DataFrame(cursor)
```

```python
best_videos = collection.find().sort([("views",-1)]).limit(10)
pd.DataFrame(best_videos)
```

## Question 5

```python
category = collection.aggregate([{"$group" : {"_id" : "$category_id", "moyenne de vues" : {"$avg" :"$views"}}}])
pd.DataFrame(category)
```

## Question 6 

```python
best_channels = collection.aggregate([{"$group":{"_id":"$channel_title","moyenne de likes":{"$avg":"$likes"}}},{"$sort" :{"avgLike": -1}},{"$limit":10}])
pd.DataFrame(best_channels)
```
